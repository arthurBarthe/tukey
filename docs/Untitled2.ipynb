```python
def tukey(g, h, z):
    return 1 / g * (np.exp(g * z) - 1) * np.exp(1 / 2 * h * z**2)
```

```python
from scipy.stats import norm
```

```python
from numpy.random import rand
n = 100000

z1, z2 = norm().rvs(n), norm().rvs(n)

g1, h1 = -0.2, 0.2
g2, h2 = 0.2, 0.05


print(g1, h1)
print(g2, h2)

t =  tukey(g2, h2, 1 / 4 * z1 +  1 / 2 * z2) + tukey(g2, h2, 1 / 2 * z1 + 1 / 4 * z2)

plt.figure()
_ = plt.hist(t, bins=500, density=True, log=False)
```
